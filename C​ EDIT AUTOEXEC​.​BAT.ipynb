So there are single spaces and spaces in groups of five, those could delimit characters and words...

```python
morse_words = string.split(' ' * 5)

def to_word(morse_word):
    return ''.join(
        [alphabet.get(char, '?') for char in morse_word.split()]
    )

words = [to_word(mw) for mw in morse_words]

words
```
